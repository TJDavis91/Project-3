```python
# Load data for Gold, Silver, and Bitcoin from CSV files
gold_data = pd.read_csv("currency_csvs/gold.csv")
silver_data = pd.read_csv("currency_csvs/silver.csv")
bitcoin_data = pd.read_csv("currency_csvs/bitcoin.csv")

# Convert the "Month_Year" column to a datetime format for each dataset
gold_data['Month_Year'] = pd.to_datetime(gold_data['Month_Year'])
silver_data['Month_Year'] = pd.to_datetime(silver_data['Month_Year'])
bitcoin_data['Month_Year'] = pd.to_datetime(bitcoin_data['Month_Year'])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot Closing Prices for Gold, Silver, and Bitcoin
ax.plot(gold_data['Month_Year'], gold_data['Closing_Price'], label='Gold', color='gold')
ax.plot(silver_data['Month_Year'], silver_data['Closing_Price'], label='Silver', color='silver')
ax.plot(bitcoin_data['Month_Year'], bitcoin_data['Closing_Price'], label='Bitcoin', color='orange')

# Set axis labels and title
ax.set_xlabel('Month_Year')
ax.set_ylabel('Closing Price')
ax.set_title("Closing Prices Over Time for Gold, Silver, and Bitcoin")

# Add a legend
ax.legend()

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

```

```python
# Load the data from the CSV file
data = pd.read_csv("Daily_Gold_Price_on_World.csv")

# Convert the "Date" column to a datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot exchange rates for selected currencies (e.g., USD, EUR, JPY)
currencies = ['US dollar (USD)', 'Euro (EUR)', 'Japanese yen (JPY)']
colors = ['blue', 'green', 'red']

for i in range(len(currencies)):
    ax.plot(data['Date'], data[currencies[i]], color=colors[i], label=currencies[i])

# Set axis labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Exchange Rate')
ax.set_title("Exchange Rates Over Time")

# Add a legend
ax.legend()

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the currency exchange rate data
data = pd.read_csv("Daily_Gold_Price_on_World.csv")

# Choose a currency column for the pivot (e.g., "US dollar (USD)")
pivot_data = data.pivot("Date", "US dollar (USD)")

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap="coolwarm", linewidths=.5, annot=True)
plt.title("Currency Exchange Rates Over Time")
plt.xlabel("Currency")
plt.ylabel("Date")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
